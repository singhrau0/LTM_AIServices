```python
import os

# Download the zip file
!wget https://github.com/singhrau0/LTM_AIServices/raw/refs/heads/main/Day%20-%208/cats_and_dogs.zip -O cats_and_dogs.zip

# Extract the zip file
!unzip -q cats_and_dogs.zip

import tensorflow as tf
import matplotlib.pyplot as plt

BATCH_SIZE = 32
IMG_SIZE = (160, 160)

# Load datasets from directories
train_dataset = tf.keras.utils.image_dataset_from_directory(
    '/content/train',
    shuffle=True,
    batch_size=BATCH_SIZE,
    image_size=IMG_SIZE)

validation_dataset = tf.keras.utils.image_dataset_from_directory(
    '/content/validation',
    shuffle=True,
    batch_size=BATCH_SIZE,
    image_size=IMG_SIZE)

# Note: User specified test images are in /content/train as well per prompt, 
# but standard practice uses a separate test folder if available.
test_dataset = tf.keras.utils.image_dataset_from_directory(
    '/content/test',
    shuffle=True,
    batch_size=BATCH_SIZE,
    image_size=IMG_SIZE)

class_names = train_dataset.class_names
print(f'Class names: {class_names}')

# Define expanded data augmentation layers
data_augmentation = tf.keras.Sequential([
  tf.keras.layers.RandomFlip('horizontal'),
  tf.keras.layers.RandomRotation(0.2),
  tf.keras.layers.RandomZoom(0.2), # Zooming
  tf.keras.layers.RandomTranslation(height_factor=0.1, width_factor=0.1) # Shifting/Shearing equivalent
])

# Preprocessing layer (rescaling pixels to [0,1])
rescale = tf.keras.layers.Rescaling(1./255)

# Visualize the new augmented images
plt.figure(figsize=(10, 10))
for images, labels in train_dataset.take(1):
    for i in range(9):
        # Apply both augmentation and rescaling for visualization
        augmented_images = data_augmentation(images)
        ax = plt.subplot(3, 3, i + 1)
        # Convert back to uint8 for plotting if needed, or keep as float if normalized
        plt.imshow(augmented_images[0].numpy().astype("uint8"))
        plt.title(class_names[labels[0]])
        plt.axis("off")
plt.suptitle("Augmented Images (with Zoom & Translation)")
plt.show()


```
