```python
tre['Joined'] = pd.to_datetime(tre['Joined'])
```

```python
import datetime as dt
```

```python
tre['joinyear'] = tre['Joined'].dt.year
tre['joinmonth'] = tre['Joined'].dt.month
tre['joinday'] = tre['Joined'].dt.day
```

```python
tre
```

```python
# i = 132.0
def valueconvertor(col):
  col = col.str.replace('€','')
  data = []
  for i in col:
    if i.endswith("M"):
      i = float(i.replace("M",''))*1000000
      data.append(i)
    elif i.endswith("K"):
      i = float(i.replace("K",''))*1000
      data.append(i)
    else:
      data.append(float(i))
  return data
```

```python
tre['Value'] = valueconvertor(tre['Value'])
tre['Wage'] = valueconvertor(tre['Wage'])
tre['Release Clause'] = valueconvertor(tre['Release Clause'])
```
