## 1. Import Libraries

```python
import re
from collections import Counter
```

## 2. Example Text

```python
text = """
Artificial intelligence is becoming useful in education, healthcare, finance, and customer support.
Students can use AI tools to revise topics, ask questions, and practice coding.
Companies use AI assistants to summarize documents, classify customer feedback, and automate repetitive work.
However, AI systems should be checked carefully because they can make mistakes.
Good AI applications need clear instructions, useful data, testing, and responsible human review.
"""
print(text)
```

## 3. Clean and Split Sentences

```python
def split_sentences(text):
    sentences = re.split(r"(?<=[.!?])\s+", text.strip())
    return [sentence.strip() for sentence in sentences if sentence.strip()]

sentences = split_sentences(text)
for number, sentence in enumerate(sentences, start=1):
    print(number, sentence)
```

## 4. Beginner Extractive Summarization

Extractive summarization selects important sentences from the original text.

```python
stopwords = {"is", "in", "and", "to", "use", "can", "they", "the", "a", "an", "of", "because"}


def clean_words(sentence):
    sentence = sentence.lower()
    sentence = re.sub(r"[^a-z\s]", " ", sentence)
    return [word for word in sentence.split() if word not in stopwords]

all_words = []
for sentence in sentences:
    all_words.extend(clean_words(sentence))

word_counts = Counter(all_words)
print(word_counts.most_common(10))
```

```python
def summarize(text, number_of_sentences=2):
    sentences = split_sentences(text)
    scores = []
    for sentence in sentences:
        words = clean_words(sentence)
        score = sum(word_counts.get(word, 0) for word in words)
        scores.append((sentence, score))

    top_sentences = sorted(scores, key=lambda item: item[1], reverse=True)[:number_of_sentences]
    selected = {sentence for sentence, score in top_sentences}
    summary = [sentence for sentence in sentences if sentence in selected]
    return " ".join(summary)

print(summarize(text, number_of_sentences=2))
```

## 5. Beginner Sentiment Analysis

```python
positive_words = {"good", "great", "useful", "clear", "helpful", "excellent", "love", "responsible"}
negative_words = {"bad", "poor", "slow", "confusing", "mistake", "mistakes", "hate", "repetitive"}


def simple_sentiment(text):
    words = clean_words(text)
    positive_score = sum(word in positive_words for word in words)
    negative_score = sum(word in negative_words for word in words)

    if positive_score > negative_score:
        label = "positive"
    elif negative_score > positive_score:
        label = "negative"
    else:
        label = "neutral"

    return {"text": text, "positive_score": positive_score, "negative_score": negative_score, "sentiment": label}

examples = [
    "The AI assistant gave a clear and useful answer.",
    "The tool was confusing and made many mistakes.",
    "The system answered the question."
]

for example in examples:
    print(simple_sentiment(example))
```

## 6. Simple Conversational AI Pattern

A chatbot usually needs user messages, memory, response logic, and fallback responses.

```python
conversation_memory = []


def simple_chatbot(user_message):
    message = user_message.lower()
    conversation_memory.append({"role": "user", "message": user_message})

    if "summary" in message or "summarize" in message:
        response = summarize(text, number_of_sentences=2)
    elif "sentiment" in message:
        response = "Send me a sentence, and I can check whether it is positive, negative, or neutral."
    elif "hello" in message or "hi" in message:
        response = "Hello. I can help with summary, sentiment, and revision questions."
    elif "memory" in message:
        response = f"I have {len(conversation_memory)} user messages in memory."
    else:
        response = "I did not understand fully. Try asking for a summary or sentiment help."

    conversation_memory.append({"role": "assistant", "message": response})
    return response

print(simple_chatbot("hello"))
print(simple_chatbot("summarize the text"))
print(simple_chatbot("show memory"))
```

## 7. Prompt Template Practice

Prompt templates help make instructions consistent.

```python
def build_prompt(task, content, output_style="short and clear"):
    prompt = f"""
Task: {task}
Output style: {output_style}
Content:
{content}
"""
    return prompt.strip()

prompt = build_prompt(
    task="Summarize this text for a beginner student",
    content=text,
    output_style="3 bullet points"
)
print(prompt)
```

## 8. Mini Project: Student Revision Helper

```python
def revision_helper(question):
    question_lower = question.lower()

    if "summarize" in question_lower:
        return summarize(text, 2)
    if "positive" in question_lower or "negative" in question_lower:
        return simple_sentiment(question)
    if "ai application" in question_lower:
        return "Common AI applications include summarization, sentiment analysis, chatbots, search, and document question answering."
    return "Please ask about summarization, sentiment, or AI applications."

print(revision_helper("What are common AI applications?"))
print(revision_helper("Summarize the paragraph"))
```
